```python
# Libraries required for the dashboard

import base64
import datetime
import io

import dash
from dash.dependencies import Input, Output, State
import dash_core_components as dcc
import dash_html_components as html
import dash_table

import pandas as pd
```

```python
app = dash.Dash()

# Defining the HTML components

app.title = 'NAME'

app.layout = html.Div(children = [
    html.Div("Tung/Korey Site",
        style = {
            'backgroundColor':'white',
            'color':'black',
            'fontFamily':'Arial, sans-serif',
            'fontSize':'50px',
            'padding':'30px',
            'textAlign':'center',
    }),
    html.Div("NAME automates the processing of ancient residue metabolomics datasets. Upload "
             "a CSV file and we will take care of the rest. The output will be a table. Thank "
             "you for processing your data with us!",
        style = {
            'backgroundColor':'white',
            'color':'black',
            'fontFamily':'Arial, sans-serif',
            'fontSize':'20px',
            'padding':'5px',
            'textAlign':'left',
    }),
    html.Div("Please Cite:",
        style = {
            'backgroundColor':'white',
            'color':'black',
            'fontFamily':'Arial, sans-serif',
            'fontSize':'15px',
            'marginTop':'30px',
            'padding':'5px',
            'textAlign':'left',
    }),
    html.Div("Nguyen, T. T., and Brownstein, K. J. (2021). NAME: a web platform for processing " 
             "and analyzing ancient residue metabolomics datasets.",
        style = {
            'backgroundColor':'white',
            'color':'black',
            'fontFamily':'Arial, sans-serif',
            'fontSize':'15px',
            'marginBottom':'30px',
            'padding':'5px',
            'textAlign':'left',
    }),
    dcc.Upload(
        id = 'upload-data',
        children = html.Div([
            'Drag and Drop, or ',
            html.A('Select a CSV File')
        ]),
        style = {
            'backgroundColor':'#f2f2f2',
            'borderRadius':'5px',
            'borderStyle':'dashed',
            'borderWidth':'2px',
            'color':'black',
            'fontFamily':'Arial, sans-serif',
            'fontSize':'20px',
            'height':'200px',
            'lineHeight':'200px',
            'textAlign':'center',
            'width':'100%',
        },
        # Allows for multiple files to be uploaded
        multiple = True
    ),
    html.Div("Results:",
        style = {
            'backgroundColor':'white',
            'color':'black',
            'fontFamily':'Arial, sans-serif',
            'fontSize':'15px',
            'marginTop':'30px',
            'padding':'5px',
            'textAlign':'left',
    }),
    html.Div(id = 'output-data-upload',
        style = {
            'padding':'10px',
    }),
])

def parse_contents(contents, filename, date):
    content_type, content_string = contents.split(',')

    decoded = base64.b64decode(content_string)
    try:
        if 'csv' in filename:
            # Assumes that the user uploaded a CSV file
            df = pd.read_csv(
                io.StringIO(decoded.decode('utf-8')))
        elif 'xls' in filename:
            # Assumes that the user uploaded an excel file
            df = pd.read_excel(io.BytesIO(decoded))
    except Exception as e:
        print(e)
        return html.Div([
            'There was an error processing this file.'
        ])

    return html.Div([
        html.H5(filename),
        html.H6(datetime.datetime.fromtimestamp(date)),

        dash_table.DataTable(
            data = df.to_dict('records'),
            columns = [{'name': i, 'id': i} for i in df.columns]
        ),
        
        # Horizontal line
        html.Hr(),

        # For debugging, display the raw contents provided by the web browser
        html.Div('Raw Content'),
        html.Pre(contents[0:200] + '...', 
        style = {
            'wordBreak':'break-all',
            'whiteSpace':'pre-wrap',
        })
    ])

@app.callback(Output('output-data-upload', 'children'),
              Input('upload-data', 'contents'),
              State('upload-data', 'filename'),
              State('upload-data', 'last_modified'))
def update_output(list_of_contents, list_of_names, list_of_dates):
    if list_of_contents is not None:
        children = [
            parse_contents(c, n, d) for c, n, d in
            zip(list_of_contents, list_of_names, list_of_dates)]
        return children

if __name__ == '__main__':
    app.run_server(debug = False)
```
