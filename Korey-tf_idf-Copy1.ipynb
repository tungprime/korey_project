```python
def first_step(data, number_of_samples, number_of_groups):
    """    
    number_of_samples: the number of different samples for each group. 
    
    number_of_groups: the number of different groups. 
    
    return the result whether we compute the sum for each group. 
    """
    
    s=number_of_samples
    g=number_of_groups
    df=pd.DataFrame()
    for i in range(number_of_groups):
        name=data.columns[i*s].split('.')[0]
        df[name]=data.iloc[:, i*s: (i+1)*s].max(axis=1)
    new_df=df[df.iloc[:,-1]==0].iloc[:, :-1]
    row_sum=new_df.sum(axis=1)
    final_sample=new_df[row_sum != 0]
    return final_sample    
    
        
```
